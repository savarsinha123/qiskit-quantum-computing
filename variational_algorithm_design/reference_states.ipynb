## Import Statements

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import TwoLocal
from qiskit.circuit.library import ZZFeatureMap
from math import pi
```

The first step of any variational algorithm is to initialize a reference state via a unitary operation. This state can take various forms as shown below.

## Default State

The default state is simply $| 0 \rangle ^{\otimes n}$, meaning that the unitary operation applied is the identity operator.

## Classical States

For the classical reference, we simply use $X$ operators to modify a subset of the zeros in the default state to 1. For example, see the circuit below which creates the state $|001\rangle$ by Qiskit convention.

```python
qc = QuantumCircuit(3)
qc.x(0)
qc.draw('mpl')
```

## Quantum Reference States

Suppose we want to create a reference state with entanglement or superposition. We can accomplish this by using Hadamard operations as follows. Consider the following circuit which creates a reference state $\frac{1}{\sqrt 2}|000\rangle + \frac{1}{\sqrt 2}|110\rangle$.

```python
qc = QuantumCircuit(3)
qc.h(1)
qc.cx(1, 2)
qc.draw('mpl')
```

## Template Circuits

There are other template circuits we can use such as the ``TwoLocal`` template from Qiskit which more easily allows for tunable parameters and entanglement.

```python
reference_circuit = TwoLocal(2, "rx", "cz", entanglement="linear", reps=1)
theta_list = [pi / 2, pi / 3, pi / 3, pi / 2]
reference_circuit = reference_circuit.assign_parameters(theta_list)
reference_circuit.decompose().draw("mpl")
```

## Application-specific Reference States

In the context of quantum machine learning, there exists a type of classifier known as a variational quantum classifier (VQC), in which training data is encoded into a quantum state with a parameterized circuit known as a feature map, where each parameter value represents a data point from the training dataset. To pass data points to this feature map, we can utilize the ``ZZFeatureMap`` reference.

```python
data = [0.2, 0.3]
zz_feature_map_reference = ZZFeatureMap(feature_dimension=2, reps=2)
zz_feature_map_reference = zz_feature_map_reference.assign_parameters(data)
zz_feature_map_reference.decompose().draw("mpl")
```
